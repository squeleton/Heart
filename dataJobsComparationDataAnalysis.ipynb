```python
# Filter data for each role
data_engineer = data[data['job_title'].str.contains('Data Engineer', case=False)]
data_scientist = data[data['job_title'].str.contains('Data Scientist', case=False)]
data_analyst = data[data['job_title'].str.contains('Data Analyst', case=False)]

# Calculate median salary and count job offers
data_engineer_median_salary = data_engineer['salary'].mean()
data_engineer_job_count = data_engineer.shape[0]

data_scientist_median_salary = data_scientist['salary'].mean()
data_scientist_job_count = data_scientist.shape[0]

data_analyst_median_salary = data_analyst['salary'].mean()
data_analyst_job_count = data_analyst.shape[0]

# Print results
print(f"Data Engineer - Median Salary: {data_engineer_median_salary}, Job Count: {data_engineer_job_count}")
print(f"Data Scientist - Median Salary: {data_scientist_median_salary}, Job Count: {data_scientist_job_count}")
print(f"Data Analyst - Median Salary: {data_analyst_median_salary}, Job Count: {data_analyst_job_count}")
```

```python
# Create a summary DataFrame
summary_df = pd.DataFrame({
    'Role': ['Data Engineer', 'Data Scientist', 'Data Analyst'],
    'Median Salary': [data_engineer_median_salary, data_scientist_median_salary, data_analyst_median_salary],
    'Job Count': [data_engineer_job_count, data_scientist_job_count, data_analyst_job_count]
})

print(summary_df)
```

```python

# Set the plot size
plt.figure(figsize=(12, 5))

# Create a bar plot for median salaries
plt.subplot(1, 2, 1)
sns.barplot(x='Role', y='Median Salary', data=summary_df, palette='viridis')
plt.title('Median Salary by Role')
plt.xlabel('Role')
plt.ylabel('Median Salary')

# Create a bar plot for job counts
plt.subplot(1, 2, 2)
sns.barplot(x='Role', y='Job Count', data=summary_df, palette='viridis')
plt.title('Job Count by Role')
plt.xlabel('Role')
plt.ylabel('Job Count')

# Show the plots
plt.tight_layout()
plt.show()
```
