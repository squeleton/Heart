```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Parkinsson disease.csv")
data.head(10)
```

```python
#Checkers
data.isnull().sum()
data.info()
data.shape
data.describe()
```

```python
data['status'].value_counts()
# Status is going to be the dependent variable
# 1 = Brain with Parkinson / 2 = Normal brain
```

```python
x = data.drop(columns = ['name','status'], axis=1) # Independet variables - status and name
y = data['status'] #Target variable(dependet)
```

```python
y.shape
```

```python
x.shape
```

```python
# let us import sklearn
from sklearn.model_selection import train_test_split
```

```python
#We split the data; 80% to the train part and 20% to the test part
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size= 0.2,random_state=2)
y_train.shape
y_test.shape
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
scala = StandardScaler()
scala.fit_transform(x_train)
scala.fit_transform(x_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'linear')
model.fit(x_train,y_train)
```

```python
prediTrain = model.predict(x_train)
prediTrain
```

```python
from sklearn.metrics import accuracy_score
accTrain = accuracy_score(y_train, prediTrain)
accTrain
```

```python
prediTest = model.predict(x_test)
accTest = accuracy_score(y_test, prediTest)
accTest
```

```python
#Predictive Model
input = (119.992,	157.302,	74.997,	0.00784,	0.00007,	0.0037,	0.00554,	0.01109,	0.04374,	0.426,	0.02182,	0.0313,	0.02971,	0.06545,	0.02211,	21.033,	0.414783,	0.815285,	-4.813031,	0.266482,	2.301442,	0.284654)
inputInArray = np.asarray(input)
reshapee = inputInArray.reshape(1,-1)
stdData = scala.transform(reshapee)
prediction = model.predict(stdData)
```
